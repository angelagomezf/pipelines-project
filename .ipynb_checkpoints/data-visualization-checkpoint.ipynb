## 2. Plotting

### HYPOTHESIS: Followers on Instagram vs. Spotify
- Does the number of followers on Instagram match the number of followes on Spotify? Let's find out.

### Instagram

```python
i = sns.barplot(x="BRAND", y="FOLLOWERS (M)", data=df)
i.set_xticklabels(i.get_xticklabels(),rotation=90)
sns.set(rc={'figure.figsize':(15,8)})
plt.show()
```

#### Key Takeaways:
- The musician with most followers is **Selena Gomez**, followed by Taylor Swift and Ariana Grande.
- The difference in followers between Taylor Swift (#2), Ariana Grande (#3) and Beyonce (#4) is minimal.
- The difference between the most followed (Selena Gomez) and the last from the list (Tyga), is substantial: 90M.

### Spotify

```python
s = sns.barplot(x="SPOTIFY NAMES", y="SPOTIFY FOLLOWERS", data=df)
s.set_xticklabels(s.get_xticklabels(),rotation=90)
sns.set(rc={'figure.figsize':(15,8)})
plt.show()
```

#### Key Takeaways:
- The musician with most followers is **Ariana Grande**, followed by Drake and Rihanna.
- The differences in followers between Ariana Grande (#1) and Drake (#2), is minimal. Whereas the difference between Drake and Rihana (#3), is more significant.
- There is a big gap between the most and the least followed from this list (Zendaya).

### Let's draw some conclusions:
1. The most followed musician on one platform **is not** the most followed on the other: On Instagram, the artist with most followers is *Selena Gomez*, whereas on Spotify, it's *Ariana Grande*.
2. Some users follow their favorite artist on **both paltforms**, as we can see in the case of Ariana Grande (she is #3 on Instagram and #1 on Spotify)
3. These discrepancies might be due to the fact that even though Spotify can be enjoyed for free, it is a **paid platform**, whereas on the other hand, Instagram is not. Moreover
