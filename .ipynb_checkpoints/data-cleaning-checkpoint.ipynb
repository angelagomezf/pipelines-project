# Data Cleaning

- First things first: choosing a dataset and cleaning it, to then proceed to enriching and analyzing it.
- For this project I have chosen "the most followed accounts on Instagram" database from Kaggle.

## 1. Loading the data

#### Let's import the libraries first

```python
import pandas as pd 
import numpy as np
import re
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

#### Now, let's load the data

```python
df = pd.read_csv('most_followed_ig.csv', encoding = 'Windows-1252')
df
```

## 2. Exploratory Analysis

```python
df.shape
```

```python
df.dtypes
```

```python
df.columns.values
```

### Renaming columns

```python
df.rename(columns = {'CATEGORIES 1':'CATEGORIES', 'CATEGORIES 2':'SUBCATEGORIES'}, inplace = True)
```

```python
df
```

### Removing special characters

```python
df.replace('æ','',regex=True, inplace=True)
```

```python
df.replace('\(.*\)','',regex=True, inplace=True)
```

```python
df
```

### Null values

```python
df.isnull().values.any()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

## 3. Plotting

#### Let's see how many different CATEGORIES are

```python
df['CATEGORIES'].value_counts()
```

The most important category is "*celebrities*", so let's dive in a little bit more.

```python
df['SUBCATEGORIES'].value_counts()
```

The most important subcategory is "*musicians*", so let's assign it to a variable so we can work better with it.

```python
my_labels = df['SUBCATEGORIES'].value_counts()
top_5 = my_labels[:5]
top_5
```

```python
my_colors = ['lightblue','lightsteelblue','silver']
my_explode = (0.1, 0, 0, 0, 0)
plt.pie(top_5, labels=top_5.index, autopct='%1.1f%%', startangle=15, shadow = True, colors=my_colors, explode=my_explode)
plt.title('SUBCATEGORIES')
plt.axis('equal')
plt.show()
```

As a result of the *exploratory analysis*, we conclude that the most relevant subcategory is **musicians**. Which is why we're going to proceed to enrich its subset, analyze it, and last but not least: draw conclusions.

## 3. Saving clean dataset

#### Saving entire dataset clean

```python
df.to_csv('most_followed_clean.csv')
```

#### Saving dataset with only musicians subcategory, to enrich it later on

```python
df_musicians = df.loc[df['SUBCATEGORIES'] == 'musicians']
df_musicians
```
